# Remover Duplicatas de InChIKeys Mantendo Maior pKd

Este notebook compara InChIKeys duplicados e mantém apenas as moléculas com o maior valor de pKd.

## 1. Importar Bibliotecas Necessárias

```python
import pandas as pd
import numpy as np
```

## 2. Configurar Caminhos dos Arquivos

**IMPORTANTE:** Altere os caminhos abaixo para cada dataset antes de executar o notebook.

```python
import os

# ========================================
# CONFIGURAÇÃO DE CAMINHOS
# Altere aqui para cada dataset
# ========================================

# Pasta onde estão os arquivos
folder_path = '/Users/francisco/Documents/Scripts/DeNovo_PfHDAC1/LigBuilder_results/Build_PfHDAC1_HIGHDIV_MAY/result/process_PfHDAC1'

# Nomes dos arquivos
input_filename = 'extracted_denovo.csv'
output_filename = 'dados_sem_duplicatas.csv'

# Separador do CSV (geralmente ',' ou ';')
csv_separator = ';'

# Construir caminhos completos
input_file = os.path.join(folder_path, input_filename)
output_file = os.path.join(folder_path, output_filename)

print("Configuração de caminhos:")
print(f"Pasta: {folder_path}")
print(f"Arquivo de entrada: {input_filename}")
print(f"Arquivo de saída: {output_filename}")
print(f"Separador: '{csv_separator}'")
```

## 3. Criar ou Carregar DataFrame

Você pode criar um DataFrame de exemplo ou carregar seus dados de um arquivo CSV.

```python
df = pd.read_csv(input_file, sep=csv_separator)
print(df.head())
```

## 4. Identificar Duplicatas de InChIKeys

```python
# Verificar quais InChIs estão duplicados
duplicados = df[df.duplicated(subset='InChI', keep=False)]
print(f"\nInChIs duplicados: {duplicados['InChI'].nunique()} únicos")
print(f"Total de linhas duplicadas: {len(duplicados)}")
print("\nMoléculas com InChIs duplicados:")
print(duplicados.sort_values('InChI'))
```

## 5. Manter Apenas Moléculas com Maior pKd

Usar `groupby` e `idxmax` para manter apenas a linha com o maior valor de pKd para cada InChIKey.

```python
# Método 1: Usando sort_values + drop_duplicates
# Ordena por pKd descendente e mantém a primeira ocorrência de cada InChI
df_final = df.sort_values('pKd', ascending=False).drop_duplicates(subset='InChI', keep='first')

print(f"\nDataFrame final: {len(df_final)} linhas")
print(df_final.sort_values('InChI'))
```

## 6. Verificar Resultado Final

Confirmar que não há mais InChIKeys duplicados e visualizar as moléculas mantidas.

```python
# Verificar se ainda há duplicatas
print("Verificação de duplicatas no DataFrame final:")
print(f"Total de InChIs únicos: {df_final['InChI'].nunique()}")
print(f"Total de linhas: {len(df_final)}")
print(f"Há duplicatas? {df_final['InChI'].duplicated().any()}")

# Comparação antes e depois
print(f"\nResumo:")
print(f"DataFrame original: {len(df)} linhas")
print(f"DataFrame final: {len(df_final)} linhas")
print(f"Linhas removidas: {len(df) - len(df_final)}")
```

```python
# Salvar o DataFrame final
df_final.to_csv(output_file, index=False)
print(f"\nDataFrame salvo em: {output_file}")
```
